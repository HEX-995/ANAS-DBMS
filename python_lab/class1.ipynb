```python
s = input()
mid = len(s)//2

if len(s) % 2 == 0:
    first, second = s[:mid], s[mid:]
else:
    first, second = s[:mid], s[mid+1:]

print("Symmetrical" if first == second else "Not Symmetrical")
print("Palindrome" if s == s[::-1] else "Not Palindrome")
```

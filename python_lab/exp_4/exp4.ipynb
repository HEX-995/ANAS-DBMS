```python
def find_max(*numbers):
    
    if(len(numbers)==0):
        return"no value entered"
    max=numbers[0]
    
    for i in numbers:
        if(i>max):
            max=i
    return max           

print(find_max(3,6,9,3,6))
print(find_max())
```

```python
def count_char(string):
    freq={}
    for i in string:
        if i in freq:
            freq[i]+=1
        else:
            freq[i]=1
    return freq
print(count_char("dilshad"))
```

```python
def power(x,n):
    if(n==1):
        return x
    half=power(x,n//2)
    if(n%2!=0):
        return half*half*x
    else:
        return half*half

print(power(2,3))
```
